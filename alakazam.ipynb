```python
import requests
from IPython.display import Image, display

class Pokemon:
    def __init__(poke, zam):
        poke.zam = zam
        poke.taught_moves = []

class Move_Tutor(Pokemon):
    def __init__(poke, zam):
        super().__init__(zam)
        poke.move_list = []
        poke.alakazam_path = "alakazam.jpg"
        poke.alakazam_stronger_path = "alakazam_stronger.jpg"
        poke.poke_api_insertion()

    def poke_api_insertion(poke):
        try:
            response = requests.get(f"https://pokeapi.co/api/v2/pokemon/{poke.zam}")
            response.raise_for_status()
            moves = response.json()["moves"]
            poke.move_list = [move["move"]["name"].upper() for move in moves]
        except requests.exceptions.RequestException as e:
            print(f"Who's That Pokemon? (ERROR) Check Spelling: {e}")
            poke.move_list = []

    def teach_moves(poke):
        poke.zam_before()
        print("\nLet's Teach Alakazam Some New Moves!")
        print("Available Moves:")
        print(poke.move_list)

        moves_to_learn = 4
        while moves_to_learn > 0:
            moves_learned = input("Type Whichever Move You Want Alakazam to Learn or 'EXIT': ")

            if moves_learned.upper() == 'EXIT':
                break

            moves_learned_upper = moves_learned.upper()
            if moves_learned_upper in poke.move_list and moves_learned_upper not in poke.taught_moves:
                poke.taught_moves.append(moves_learned_upper)
                print(f"{poke.zam.capitalize()} Learned {moves_learned_upper}!")
                moves_to_learn -= 1
            elif moves_learned_upper in poke.taught_moves:
                print(f"{poke.zam.capitalize()} Already Knows {moves_learned_upper}.")
            else:
                print(f"(ERROR) Check Spelling and Capitalization for {moves_learned_upper}")

        poke.display_final_state()

    def zam_before(poke):
        display(Image())
        print(f"Pokemon: {poke.zam.capitalize()}")

    def display_final_state(poke):
        print(f"\n{poke.zam.capitalize()}'s Final Moves: {poke.taught_moves}")
        display(Image(filename=poke.alakazam_stronger_path))


if __name__ == "__main__":
    alakazam = Move_Tutor("alakazam")
    alakazam.teach_moves()
```
